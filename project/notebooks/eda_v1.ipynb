```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager

data = {
    'Имя': ['Анна', 'Борис', 'Виктория', 'Григорий'],
    'Возраст': [25, 30, 35, 40],
    'Город': ['Москва', 'Санкт-Петербург', 'Новосибирск', 'Екатеринбург'],
    'Зарплата': [70000, 80000, 90000, 100000]
}

df = pd.DataFrame(data)

# Настройка шрифта
plt.rcParams['font.family'] = 'DejaVu Sans'  # Можно заменить на 'Arial', 'Times New Roman'
plt.rcParams['font.size'] = 11

# Создаем фигуру
fig, ax = plt.subplots(figsize=(10, 4), facecolor='lightblue')
ax.axis('tight')
ax.axis('off')
ax.set_facecolor('lightblue')

# Подготовка цветов для ячеек (5 строк: 1 заголовок + 4 строки данных)
num_rows = len(df) + 1  # +1 для строки заголовков
num_cols = len(df.columns)

# Создаем таблицу БЕЗ предустановленных цветов в аргументах
table = ax.table(cellText=df.values,
                 colLabels=df.columns,
                 cellLoc='center',
                 loc='center')

# Теперь настраиваем цвета ПОСЛЕ создания таблицы
# 1. Цвет заголовков
for i in range(num_cols):
    cell = table[(0, i)]  # Строка 0 - заголовки
    cell.set_facecolor('navy')
    cell.set_text_props(color='white', weight='bold')
    cell.set_height(0.15)  # Высота ячейки

# 2. Цвет строк данных (чередование для лучшей читаемости)
for i in range(1, num_rows):  # Начинаем с 1, пропускаем заголовки
    for j in range(num_cols):
        cell = table[(i, j)]
        # Чередующиеся цвета строк
        if i % 2 == 1:  # Нечетные строки (после заголовка)
            cell.set_facecolor('#f0f0f0')  # Светло-серый
        else:  # Четные строки
            cell.set_facecolor('#ffffff')  # Белый

# 3. Настройка границ
for key, cell in table.get_celld().items():
    cell.set_edgecolor('gray')
    cell.set_linewidth(0.5)

# 4. Настройка размера ячеек
table.scale(1.2, 1.8)  # (ширина, высота)

# 5. Дополнительная настройка шрифта для всей таблицы
table.auto_set_font_size(False)
table.set_fontsize(11)

# Сохраняем
plt.savefig('таблица_данных_matplotlib.jpg', 
            bbox_inches='tight', 
            dpi=300,
            facecolor=fig.get_facecolor(),
            edgecolor='none')

plt.close()
print("Таблица сохранена как 'таблица_данных_matplotlib.jpg'")
```
