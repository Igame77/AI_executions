```python
def dublicate_with_noise(df, noise_level = 0.01):
    df_minor = df[df['target']==1][:100]

    df_res_minor = []
    for i in range(95//5 * 25000 // (int(5/95 * 25000) * 60) - 2):
        df_noise = df_minor.copy()
        for col in df.columns:
            if col not in ['id', 'target']:
                noise = np.random.normal(0, noise_level, size = df_noise[col].shape)
                df_noise[col]+=noise
                df_res_minor.append(df_noise)
    df_res = pd.concat([df, *df_res_minor], ignore_index=True)
    return df_res

df = dublicate_with_noise(df)
```

Для того, чтобы уровнять количество данных, мы воспользовались дублированием строк, где target=1, с добавлением шума в каждый столбец отдельно, тем самым восполняя дисбаланс, не смещая данные.

```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix

metrics_data = []

def calculate_metrics(y_true, y_pred, y_proba, model_name):
    metrics = {
        'Model': model_name,
        'Accuracy': accuracy_score(y_true, y_pred),
        'Precision': precision_score(y_true, y_pred),
        'Recall': recall_score(y_true, y_pred),
        'F1-Score': f1_score(y_true, y_pred),
        'ROC-AUC': roc_auc_score(y_true, y_proba) if y_proba is not None else None
    }
    
    cm = confusion_matrix(y_true, y_pred)
    metrics['TN'], metrics['FP'], metrics['FN'], metrics['TP'] = cm.ravel()
    
    return metrics

dummy_pred = baselineDummy.predict(X_test)
dummy_proba = baselineDummy.predict_proba(X_test)[:, 1]
dummy_metrics = calculate_metrics(y_test, dummy_pred, dummy_proba, 'DummyClassifier')
metrics_data.append(dummy_metrics)

lr_pred = baselineLine.predict(X_test)
lr_proba = baselineLine.predict_proba(X_test)[:, 1]
lr_metrics = calculate_metrics(y_test, lr_pred, lr_proba, 'LogisticRegression')
metrics_data.append(lr_metrics)

dt_pred = tree_model.predict(X_test)
dt_proba = tree_model.predict_proba(X_test)[:,1]
dt_metrics = calculate_metrics(y_test, dt_pred, dt_proba, 'DecisionTreeClassifier')
metrics_data.append(dt_metrics)

rf_pred = forest_model.predict(X_test)
rf_proba = forest_model.predict_proba(X_test)[:,1]
rf_metrics = calculate_metrics(y_test, rf_pred, lr_pred, 'RandomForest')
metrics_data.append(rf_metrics)

gb_pred = gradient_model.predict(X_test)
gb_proba = gradient_model.predict_proba(X_test)[:,1]
gb_metrics = calculate_metrics(y_test, gb_pred, gb_proba, 'GradientBoosting')
metrics_data.append(gb_metrics)

metrics_df = pd.DataFrame(metrics_data)

print("Метрики качества моделей:")
print("=" * 80)
print(metrics_df.to_string(index=False))
print("=" * 80)

metrics_df.to_csv('artifacts/metrics_test.csv', index=False)
print("\nМетрики сохранены в файл: model_metrics.csv")

```
